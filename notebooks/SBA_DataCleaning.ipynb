### 1. Target Variable Definition (`target`)
The raw dataset contains 9 distinct loan statuses in the `loanstatus` column. To build a binary classification model, we must define what constitutes a default:

* **Default (1):** `CHGOFF` (Charged Off), `LIQUID` (Liquidation of collateral), and `SOLDCO` (Sold after charge-off).
* **Non-Default (0):** `P I F` (Paid in Full), `CLSLN` (Closed normally), `SOLDNC` (Sold, performing), and `PURCH(NOT C/O)` (Purchased, performing).
* **Dropped:** `CANCLD` (Cancelled before disbursement - no credit event occurred), `CURR` (Active), `DELINQ` / `PSTDUE` / `DEFERD` (Active but late/deferred - final outcome is right-censored/unknown).

```python
# Define status groups
default_statuses = ['CHGOFF', 'LIQUID', 'SOLDCO']
non_default_statuses = ['P I F', 'CLSLN', 'SOLDNC', 'PURCH(NOT C/O)']

# Keep only rows with a known outcome
valid_statuses = default_statuses + non_default_statuses
df_clean = df[df['loanstatus'].isin(valid_statuses)].copy()

# Create binary target
df_clean['target'] = df_clean['loanstatus'].isin(default_statuses).astype(int)

# Verify
print(f"Rows before: {len(df):,}")
print(f"Rows after:  {len(df_clean):,}")
print(f"Dropped:     {len(df) - len(df_clean):,}\n")
print(df_clean['target'].value_counts())
print(f"\nDefault rate: {df_clean['target'].mean():.2%}")
```

### 2. Systematic Missingness & Temporal Filtering (Post-2010)
An analysis of the `initialinterestrate` column revealed that **61% of all records are missing**. 

However, grouping by the approval year shows this missingness is systematic rather than random:
* **Pre-2010:** ~95% to 100% of interest rate data is missing.
* **2010–Present:** 0% of interest rate data is missing.

**Decision (Option B):** 
Instead of dropping this highly predictive feature or performing massive imputation on older records, we restrict our dataset to loans approved from **2010 onwards**. This ensures:
1. Complete, high-quality interest rate data.
2. Relevance to the modern, post-2008 Financial Crisis credit environment.
3. Adequate sample size for training machine learning models.

```python
# Filter for loans approved in 2010 or later
df_modern = df_clean[df_clean['approvalfy'] >= 2010].copy()

# Print the size and new default rate
print(f"Modern dataset size (2010-present): {len(df_modern):,}")
print(f"Default rate in modern dataset: {df_modern['target'].mean():.2%}")
```

NAICS codes tells us the business belongs to which industry(mainly the first 2 digits)

```python
# 1. Clean NAICS codes: extract first 2 digits
df_modern['naics_sector'] = df_modern['naicscode'].astype(str).str[:2]
# Mark anything that isn't a 2-digit number as 'Unknown'
df_modern['naics_sector'] = df_modern['naics_sector'].where(df_modern['naics_sector'].str.isnumeric(), 'Unknown')
```

```python
df_modern['businesstype'] = df_modern['businesstype'].fillna('UNKNOWN')
```

```python
age_mapping = {
    'Existing, 5 or more years': 'Existing',
    'Existing or more than 2 years old': 'Existing',
    'Less than 5 years old but at least 4': 'Existing',
    'Less than 4 years old but at least 3': 'Existing',
    
    'Startup, Loan Funds will Open Business': 'New',
    'New, Less than 1 Year old': 'New',
    'Less than 3 years old but at least 2': 'New',
    'New Business or 2 years or less': 'New',
    'Change of Ownership': 'New',
    
    'Unanswered': 'Unknown',
}
df_modern['business_age_group'] = df_modern['businessage'].map(age_mapping).fillna('Unknown')
```

```python
cols_to_check = [
    'borrstate', 
    'bankstate', 
    'bankname', 
    'franchisecode', 
    'fixedorvariableinterestind', 
    'collateralind', 
    'jobssupported'
]

for col in cols_to_check:
    print(f"\n{col}")
    print(f"Sample: {df_modern[col].head(3).tolist()}")
    print(f"Nulls: {df_modern[col].isna().sum():,}")
    print(f"Unique count: {df_modern[col].nunique()}")
    if df_modern[col].nunique() < 10:
        print(df_modern[col].value_counts(dropna=False))
```

```python

df_modern['collateralind'] = df_modern['collateralind'].map({'TRUE': 1, 'FALSE': 0})


df_modern['is_variable_rate'] = df_modern['fixedorvariableinterestind'].map({'V': 1, 'F': 0})


df_modern['jobssupported'] = pd.to_numeric(df_modern['jobssupported'], errors='coerce').fillna(0)


df_modern['bankstate'] = df_modern['bankstate'].fillna('UNKNOWN')
df_modern['bankname'] = df_modern['bankname'].fillna('UNKNOWN')


df_modern['is_franchise'] = (~df_modern['franchisecode'].isna() & ~df_modern['franchisecode'].isin(['0', '1'])).astype(int)



print(f"Collateral indicator distribution:\n{df_modern['collateralind'].value_counts()}\n")
print(f"Variable rate flag distribution:\n{df_modern['is_variable_rate'].value_counts()}\n")
print(f"Franchise flag distribution:\n{df_modern['is_franchise'].value_counts()}\n")
print(f"Average jobs supported: {df_modern['jobssupported'].mean():.2f}")
```
